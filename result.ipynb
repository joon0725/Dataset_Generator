```python
import numpy as np
import cv2
import os
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras import layers, models
from tensorflow.keras import optimizers
```

```python
adam = optimizers.Adam(learning_rate=1e-4)
ep = 200
```

```python
(train_data, train_label), (test_data, test_label) = mnist.load_data()
train_data, test_data = train_data/255.0, test_data/255.0
```

```python
fake_label = [7, 0, 8,0,0,8,3,7,7,3,6,3,9,9,3,1,0,2,7,5,7,9,1,3,2,0,8,0,3,3,9,8,1,3,0,2,1,1,9,9,0,4,0,6,6,3,9,3,7,2,3,0,4,3,5,1,3,9,9,9,3,2,2,9,1,0,7,7,2,1,8,4,1,9,8,3,3,9,3,1,1,1,6,0,5,0,0,6,0,3,3,6,3,6,5,6,3,4,9,7,]
fake_label = np.array(fake_label)
```

```python
p = "/content/data/"
fake_data = []
```

```python
for i in sorted(os.listdir(p)):
  img = cv2.imread(os.path.join(p, i), cv2.IMREAD_GRAYSCALE)
  img = 255-img
  img2 = cv2.resize(img, dsize=(28, 28), interpolation=cv2.INTER_AREA)
  fake_data.append(np.array(img2))
```

```python
fake_data = np.array(fake_data)/255.0
```

```python
fake_data.shape
```

```python
train_data.shape
```

```python
tf.random.set_seed(42)
model_less = models.Sequential([
                           layers.Conv2D(10, kernel_size=3, activation='relu', input_shape=(28, 28, 1)),
                           layers.MaxPool2D((2, 2)),
                           layers.Dropout(0.2),
                           layers.Conv2D(20, kernel_size=3, activation='relu'),
                           layers.MaxPool2D((2, 2)),
                           layers.Dropout(0.2),
                           layers.Conv2D(30, kernel_size=3, activation='relu'),
                           layers.Flatten(),
                           layers.Dense(100, activation='relu'),
                           layers.Dense(10, activation='softmax')
])
```

```python
model_less.compile(optimizer=adam, loss='sparse_categorical_crossentropy', metrics=['acc'])
```

```python
history_less = model_less.fit(train_data[:100], train_label[:100], batch_size=128, epochs=ep)
```

```python
model_less.evaluate(test_data, test_label)
```

```python
tf.random.set_seed(42)

model_ori = models.Sequential([
                           layers.Conv2D(10, kernel_size=3, activation='relu', input_shape=(28, 28, 1)),
                           layers.MaxPool2D((2, 2)),
                           layers.Dropout(0.2),
                           layers.Conv2D(20, kernel_size=3, activation='relu'),
                           layers.MaxPool2D((2, 2)),
                           layers.Dropout(0.2),
                           layers.Conv2D(30, kernel_size=3, activation='relu'),
                           layers.Flatten(),
                           layers.Dense(100, activation='relu'),
                           layers.Dense(10, activation='softmax')
])
```

```python
model_ori.compile(optimizer=adam, loss='sparse_categorical_crossentropy', metrics=['acc'])
```

```python
model_ori.fit(train_data[:200], train_label[:200], batch_size=128, epochs=ep)
```

```python
model_ori.evaluate(test_data, test_label)
```
